# Groupby Function

```python
# emp = pd.read_csv('/content/Employees.csv')
# Create a dummy DataFrame to simulate Employees.csv
data_emp = {
    'Name': ['Asha', 'Harsh', 'Sourav', 'Riya', 'Hritik', 'Shivansh', 'Rohan', 'Akash', 'Soumya', 'Kartik'],
    'Department': ['Administration', 'Marketing', 'Technical', 'Technical', 'Marketing', 'Administration', 'Technical', 'Marketing', 'Technical', 'Administration'],
    'Employment Type': ['Full-time Employee', 'Intern', 'Intern', 'Part-time Employee', 'Part-time Employee', 'Full-time Employee', 'Full-time Employee', 'Intern', 'Intern', 'Full-time Employee'],
    'Salary': [120000, 50000, 70000, 70000, 55000, 120000, 125000, 60000, 50000, 120000],
    'Years of Experience': [5, 1, 2, 3, 4, 7, 6, 2, 1, 6]
}

emp = pd.DataFrame(data_emp)

display(emp)
```

```python
emp
```

```python
grouped = emp.groupby('Department')

tech = grouped.get_group('Technical')
market = grouped.get_group('Marketing')
admin = grouped.get_group('Administration')
```

**Creating New Columns**

```python
# Creating New Columns

data['City'] = ['Delhi', 'Bangalore', 'Chennai', 'Patna', 'Pune', 'Pune', 'Mumbai']
```

```python
data
```

**Renaming Columns**

```python
data.columns = ['First Name','Current Age','College University','Location']
```

```python
data
```

**Droping Columns**

```python
data.drop(['Current Age'], axis = 1)           # Temporary Droping Syntax
```

```python
data
```

```python
data.drop(['Current Age'], axis = 1, inplace = True)           # Permanent Droping Syntax
```

```python
data
```

```python
data
```

**Apply Method**

```python
# Creating Marks Column

data['Marks'] = [201, 210, 195, 183, 225, 203, 190]
```

```python
data
```

```python
# Apply Method

data['Marks'] = data['Marks'].apply(lambda x: x/5)
```

```python
data
```

```python
# Renaming Column

data.columns = ['First Name','College University','Location', 'Marks/Subject']
```

```python
data
```

**Sort Index**

```python
data.index = [10,28,64,7,1,6,15]          # Assigning Index
```

```python
data
```

```python
data.sort_index(ascending = True)                      # Default Statement : Ascending = True
```

```python
data.sort_index(ascending = False)
```

**Sort Values**

```python
data.sort_values(by = 'Marks/Subject', ascending = False)
```
